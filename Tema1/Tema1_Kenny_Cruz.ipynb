## 1. Crear una función Python que calcule el mínimo común múltiple de 2 números.

```python
# Declaramos la función mmc que va recibir 2 parametros, que serán número1 y número2.
def minimo_comun_multiple(numero1,numero2):
    
    # Validación para que los valores ingresados no sean menor a 1.
    if numero1 <1 or numero2 <1:
        print("El número ingresado no puede ser menor a 1. vuelve a intentar")
    else:
        # Condicionales if,else para capturar el numero mayor y guardarlo en la variable mayor.
        if numero1 > numero2:
            mayor = numero1
        else:
            mayor = numero2
        # bucle donde compruebo la divisibilidad del numero mayor con los valores ingresados.   
        while(mayor % numero1 != 0) or (mayor % numero2 != 0):
            mayor += 1
        print("El Mínimo Común Múltiple de los números " + str(numero1) +"," + str(numero2) + " es: "+ str(mayor))          
        
# Ejecutamos la función mcd
minimo_comun_multiple(15,70)
```

## 3. Basándote en datos aleatorios crea un gráfico de tarta (pie chart) con Mapplotlib.

```python
# Importamos la libreria matplotlib,
import matplotlib.pyplot as plt

# Declaramos nuestras variables
titulos = [15, 14, 13, 11, 3]
equipos = ['Barcelona', 'Emelec', 'El Nacional', 'Liga de Quito', 'Delfín, Cuenca y Olmedo']

# Configuración visuales del gráfico.
plt.pie(titulos, labels=equipos,
        autopct='%1.1f%%', shadow=True, startangle=90)

# Título del Gráfico
plt.title('Campeones Nacionales de la Serie A de Ecuador')

 # Función que dibuja el gráfico.
plt.show() 
```

## 5. Explica cómo combinar 4 gráficos en una única figura(subplot).

Para combinar 4 gráficos en una sola figura utilizamos la función SUBPLOTS, la cual crea los gráficos que le indiquemos.   
Ejemplo:

```python
#Importamos las librerias pylab y numpy.
import pylab as plt
import numpy as np

# Definimos las variables y valores de las mismas.
w = np.random.randn(50)
x = np.arange(0,10,0.1)
y = [2,43,56,31,14]
z = x*np.cos(x)

fig, ax = plt.subplots(2,2)  # Indicamos que el número de cuadrantes a crear, dentro del parentesis en este caso 4 cuadrantes.
fig.set_size_inches(8,6)     # Indicamos el tamaño de los cuadrantes.
ax[0,0].plot(z)              # ax[0,1] indicamos la posición del cuadrante y el valor de la variable para generar el gráfico.
ax[0,1].plot(y, 'g')
ax[1,0].plot(x, 'r')
ax[1,1].plot(w, 'y')
plt.show()                   # Ejecuta la función para graficar las figuras.
```
